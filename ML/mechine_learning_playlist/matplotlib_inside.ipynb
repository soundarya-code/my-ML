```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
a=np.arange(0,10)
b=np.arange(20,30)
```

```python
x=np.arange(40,50)
y=np.arange(100,110)
```

```python
plt.plot(a,b,'r*--',label="numpy matplolib")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('the numpy Graph')
plt.xticks(a)
plt.yticks(b)
plt.show()
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.title('my first graph!!')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo-')
plt.subplot(2,2,4)
plt.plot(x,y,'yo--')
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,2,1)
plt.title('sin wave')
plt.plot(x,y_sin,'b-')

plt.subplot(2,2,2)
plt.title('cos wave')
plt.plot(x,y_cos,'r--')
```

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [9,15,25] 
plt.bar(x,y,color='r')
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=np.array([3,7,9])
y=np.array([6,10,20])
plt.hist(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# plt.xticks(x)
plt.title('Histogram')
plt.show()
```

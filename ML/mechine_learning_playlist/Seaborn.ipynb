# Distribution plots

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

# Correlation with heatmap

```python
df.corr()
# Caution!! the corr() only for interger values
```

```python
sns.heatmap(df.corr())
```

## JOINPLOT
Bivariate Distribution is used to determine the relation between two variables. This mainly deals with relationship between two variables and how one variable is behaving with respect to the other

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,hue='size')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
p = sns.load_dataset("penguins")
```

```python
p.head()
```

```python
color_palette=sns.color_palette()
sns.jointplot(x='sex',y='body_mass_g',data=p,palette="husl")
```

# Pairplot
A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(p,palette="husl",hue='sex')
```

# Distplot

Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(p)
```

```python
sns.displot(p['body_mass_g'],rug=True)
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

```python
sns.distplot(p['body_mass_g'],kde=True,bins=10)
```

# Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

boxplot
violinplot
countplot
bar plot

```python
sns.countplot('body_mass_g',data=p[0:10:2])
```

```python
sns.countplot('sex',data=p)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

# Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='bill_length_mm',y='body_mass_g',data=p[0:2],palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[i*2 for i in range(1,10)]
y=[i**2 for i in range(1,10)]
```

```python
# plt.plot(x,y,label='square function',color='red',linewidth=2,marker='o',linestyle='--',markersize=10,markeredgecolor='blue')
# fmt='[color][marker][line]'
# plt.plot(x,y,'b*-',label='2x')
plt.figure(figsize=(10,5),dpi=300)
x2=np.arange(0,30,0.5)
plt.plot(x,y,'b*--',label='2^x')
plt.plot(x2[:6],x2[:6]**2,'r',label='X^2')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.title('Our first graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# plt.xticks([0,1,2,3,4,5,6,0.5],color='blue')
# plt.yticks([0.7,10,11,12,13,14,15,16],color='green')
plt.legend()
plt.show()
```

# Real world example

```python
import pandas as pd
df=pd.read_csv('gas_prices.csv')
```

```python
df.info
```

```python
df.head(3)
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.plot(df['Year'],df['USA'],'r--')
```

```python
plt.plot(df['Year'],df['USA'],'r--')
plt.plot(df['Year'],df['Mexico'],'b*-')

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.figure(figsize=(20,10))
plt.plot(df.Year, df.USA, 'b.-', label='United States')
plt.plot(df.Year, df.Canada, 'r.-')
plt.plot(df.Year, df['South Korea'], 'g.-')
plt.plot(df.Year, df.Australia, 'y.-')
plt.xlabel('Year')
plt.xticks(df.Year[::3].tolist()+[2011])
plt.ylabel('$US Dollers')
plt.legend()
```

```python
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in df:
    if country in countries_to_look_at:
        plt.plot(df.Year, df[country], marker='.')
plt.show()
```

### Histogram

```python
bins=[10,20,30,40,50,60,70,80,90]
plt.figure(figsize=(10,4))
plt.hist(df.Year,color='red')
# plt.xticks(bins)
plt.xlabel('no of country')
plt.ylabel('Year')
plt.show()
```
